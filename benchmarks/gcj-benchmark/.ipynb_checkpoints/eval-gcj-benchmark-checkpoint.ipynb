# Evaluation of doLLy in Google code jam

```python
# Import Core libraries
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import seaborn as sns
from sklearn.metrics import auc

# Graphics
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
mpl.rcParams['figure.figsize'] = (8,6)
mpl.rcParams['font.family'] = 'SF Mono'
mpl.rcParams.update({'font.size': 12})
mpl.rcParams['axes.titlepad'] = 20 
mpl.rcParams['axes.titlesize'] = 12
mpl.rcParams['axes.labelpad'] = 10

# Seaborn settings
sns.set(font_scale = 1.25)
sns.set_style('ticks', {'font.family' : 'Fira Code', 
                        'axes.grid' : True, 
                        'grid.linestyle': 'dotted',
                        'grid.color': '0.8'})
sns.axes_style({'axes.grid' : 'True'})
sns.set_palette(sns.color_palette(['Black']))
```

```python
import os
import glob
import itertools

THRESHOLDS = ['30', '40', '50']
OPT = ['O0', 'O1', 'O2']

# create the ground truth file
files = map(os.path.basename, glob.glob('sourcecode/*.c'))
pairs = pd.DataFrame(list(itertools.combinations(files, 2)), columns=['filename0', 'filename1'])
def is_clone(row):
    pset0 = row['filename0'].split('_')[0]
    pset1 = row['filename1'].split('_')[0]
    if pset0 == pset1:
        return True
    else:
        return False
pairs['clone'] = pairs.apply(is_clone, axis=1)


# parse the doLLy csv files
def parse_doLLy_csv(pairs, threshold, opt):    
    doLLy_pairs = pd.read_csv(f'clone-pairs/doLLy/gcj-doLLy_{threshold}_{opt}.csv')
    doLLy_pairs[f'doLLy_{threshold}_{opt}'] = True
    doLLy_pairs = doLLy_pairs.rename(columns={'filename': 'filename0', 'filename.1': 'filename1'})
    doLLy_pairs = doLLy_pairs[['filename0', 'filename1', f'doLLy_{threshold}_{opt}']]
    doLLy_pairs = doLLy_pairs.drop_duplicates()
    doLLy_pairs_rev = doLLy_pairs.rename(columns={'filename0': 'filename1', 'filename1': 'filename0'})
    doLLy_pairs = doLLy_pairs.append(doLLy_pairs_rev, sort=False)
    pairs = pd.merge(pairs, doLLy_pairs, how='left', on=['filename0', 'filename1'], validate='1:m')
    pairs = pairs.fillna(False)
    return pairs

for threshold in THRESHOLDS:
    for opt in OPT:
        pairs = parse_doLLy_csv(pairs, threshold, opt)


# parse NiCad clone pairs
import xml.etree.ElementTree as ET

def parse_NiCad_xml(pairs, threshold):
    root = ET.parse(f'clone-pairs/NiCad/gcj_functions-blind-clones-{threshold}.xml').getroot()
    NiCad = []
    for child in root:
        if child.tag == 'clone':
            pair = []
            for source in child:
                file = source.get('file')
                file = os.path.basename(file)
                file = file[:-8]
                pair.append(file)
            NiCad.append(pair)
    NiCad_pairs = pd.DataFrame(NiCad, columns=['filename0', 'filename1']).drop_duplicates()
    NiCad_pairs_rev = pd.DataFrame(NiCad, columns=['filename1', 'filename0']).drop_duplicates()
    NiCad_pairs[f'NiCad-{threshold}'] = True
    NiCad_pairs_rev[f'NiCad-{threshold}'] = True
    NiCad_pairs = NiCad_pairs.append(NiCad_pairs_rev, sort=False)
    pairs = pd.merge(pairs, NiCad_pairs[['filename0', 'filename1', f'NiCad-{threshold}']], how='left', on=['filename0', 'filename1'], validate='1:m')
    pairs = pairs.fillna(False)
    return pairs
pairs = parse_NiCad_xml(pairs, '0.30')
pairs = parse_NiCad_xml(pairs, '0.40')
pairs = parse_NiCad_xml(pairs, '0.50')


# parse CCCD clone pairs
import re

CCCD_pairs = pd.read_csv('clone-pairs/CCCD/gcj_comparisionReport.csv')
def parse_CCCD_files(files):
    _, file0, file1 = re.split(r'gcj\.(?=[\w])', files)
    file0 = file0.split('..')[0]
    file1 = file1.split('..')[0]
    return pd.Series([file0 + '.c', file1 + '.c'], index=['filename0', 'filename1'])
    

CCCD_pairs = CCCD_pairs.join(CCCD_pairs['Files'].apply(parse_CCCD_files))
CCCD_pairs = CCCD_pairs.drop('Files', axis=1)

CCCD_pairs = CCCD_pairs.groupby(['filename0', 'filename1']).min()
CCCD_pairs = pd.DataFrame(CCCD_pairs.to_records())
CCCD_pairs = CCCD_pairs.append(CCCD_pairs.rename(columns={"filename0": "filename1", "filename1": "filename0"}), sort=False).drop_duplicates()
CCCD_pairs['CCCD-0.30'] = CCCD_pairs['LevenDistance'] <= 30.0
CCCD_pairs['CCCD-0.40'] = CCCD_pairs['LevenDistance'] <= 40.0
CCCD_pairs['CCCD-0.50'] = CCCD_pairs['LevenDistance'] <= 50.0
pairs = pd.merge(pairs, CCCD_pairs[['CCCD-0.30', 'CCCD-0.40', 'CCCD-0.50', 'filename0', 'filename1']], how='left', on=['filename0', 'filename1'], validate='1:m').drop_duplicates(subset=['filename0','filename1'])


# parse MeCC clone pairs
def parse_MeCC(pairs, threshold):
    with open(f'clone-pairs/MeCC/MeCC-{threshold}.txt') as file:
        clone_classes = []
        clone_class = []
        for line in file:
            if line.startswith('=='):
                if clone_class:
                    clone_classes.append(clone_class)
                clone_class = []
            if line.startswith('   function : '):
                function = re.search('@ (\S*)', line).group(1)
                clone_class.append(function)
    result = []
    for clone_class in clone_classes:
        clone_pairs  = itertools.combinations(clone_class, 2)
        for pair in clone_pairs:
            result.append([pair[0], pair[1], True])
    MeCC_pairs = pd.DataFrame(data=result, columns=['filename0', 'filename1', f'MeCC-{threshold}']).drop_duplicates()
    MeCC_pairs = MeCC_pairs.append(MeCC_pairs.rename(columns={"filename0": "filename1", "filename1": "filename0"}), sort=False).drop_duplicates()
    pairs = pd.merge(pairs, MeCC_pairs, how='left', on=['filename0', 'filename1'], validate='1:m').drop_duplicates(subset=['filename0', 'filename1'])
    return pairs.fillna(False)
    

pairs = parse_MeCC(pairs, '0.30')
pairs = parse_MeCC(pairs, '0.40')
pairs = parse_MeCC(pairs, '0.50')
```

```python
def confusion_matrix(tool, pairs):
    tp = len(pairs[pairs[tool] & pairs['clone']])
    fp = len(pairs[~pairs['clone'] & pairs[tool]])
    tn = len(pairs[~pairs['clone'] & ~pairs[tool]])
    fn = len(pairs[pairs['clone'] & ~pairs[tool]])

    if fn == 0:
        recall = 1.0
    else:
        recall = tp / (tp + fn)
    if fp == 0:
        precision = 1.0
    else:
        precision = tp / (tp + fp)
    if (precision + recall) == 0:
        f1 = 0.0
    else:
        f1 = 2 * (precision * recall) / (precision + recall)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    return [precision, recall, f1, tp, fp, tn, fn, accuracy]
```

```python
def stats(pairs):
    doLLy_indices = []
    doLLy_confusion_matrices = []
    for threshold in THRESHOLDS:
        for opt in OPT:
            index = f'doLLy_{threshold}_{opt}'
            doLLy_indices.append(index)
            doLLy_confusion_matrices.append(confusion_matrix(index, pairs))


    results = doLLy_confusion_matrices + [
        confusion_matrix('NiCad-0.30', pairs),
        confusion_matrix('NiCad-0.40', pairs),
        confusion_matrix('NiCad-0.50', pairs),
        confusion_matrix('CCCD-0.30', pairs),
        confusion_matrix('CCCD-0.40', pairs),
        confusion_matrix('CCCD-0.50', pairs),
        confusion_matrix('MeCC-0.30', pairs),
        confusion_matrix('MeCC-0.40', pairs),
        confusion_matrix('MeCC-0.50', pairs),
    ]
    results = pd.DataFrame(results, columns=['precision', 'recall', 'f1', 'TP', 'FP', 'TN', 'FN', 'accuracy'],
                          index=doLLy_indices + [
                                 'NiCad-0.30', 'NiCad-0.40', 'NiCad-0.50',
                                 'CCCD-0.30', 'CCCD-0.40', 'CCCD-0.50',
                                 'MeCC-0.30', 'MeCC-0.40', 'MeCC-0.50'
                                ])
    results.index.name = 'tool'
    return results.sort_values('f1', ascending=False)

def roc_plot(stats, tool):
    roc_stats = stats[stats.index.str.contains(tool)].sort_values('tool')
    tpr = roc_stats['recall']
    tpr.at['doLLy_100_01'] = 1.0
    fpr = (roc_stats['FP'] / (roc_stats['FP'] + roc_stats['TN']))
    fpr.at['doLLy_100_01'] = 1.0
    auc_val = auc(fpr, tpr)

    fig = plt.figure(figsize=[8, 8])
    plt.plot(fpr, tpr, lw=2, label='ROC curve (area = %0.2f)' % auc_val)
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC of doLLy-O1 on GCJ benchmark')
    plt.xlim([-0.05, 1.0])
    plt.ylim([0.0, 1.05])
    plt.legend(loc="lower right")
    plt.show()
    return fig
```

## Evaluation

```python
pairs = pairs.replace(np.nan, False)
results = stats(pairs)
results.sort_values('f1', ascending=False)
results.round(3)
```
